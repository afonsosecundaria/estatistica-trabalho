```python
import pandas as pd
tabela = pd.read_csv('arquivo/partidas_extracao_ofc.csv', sep=',')
display(tabela)
```

```python

import matplotlib.pyplot as plt

tabela['tipo'] = tabela['tipo'].astype(str).str.strip().str.title()

contagem_tipos = tabela['tipo'].value_counts()

plt.figure(figsize=(6,6))
plt.pie(
    contagem_tipos,
    labels=contagem_tipos.index,
    autopct='%1.1f%%',
    startangle=90,
    counterclock=False
)
plt.title("Distribuição das Partidas: Players vs Bots")
plt.show()
```

```python
tabela['tipo'] = tabela['tipo'].astype(str).str.strip().str.title()
tabela['resultado'] = tabela['resultado'].astype(str).str.upper().str.strip()

def calcular_percentuais(df):
    total = len(df)
    if total == 0:
        return {'W': 0, 'L': 0}
    contagem = df['resultado'].value_counts()
    return {
        'W': round((contagem.get('W', 0) / total) * 100, 1),
        'L': round((contagem.get('L', 0) / total) * 100, 1),
    }


percentual_player = calcular_percentuais(tabela[tabela['tipo'] == 'Player'])
percentual_bot = calcular_percentuais(tabela[tabela['tipo'] == 'Bot'])
percentual_total = calcular_percentuais(tabela)


print("Percentuais de Resultados:")
print("Player:", percentual_player)
print("Bot:", percentual_bot)
print("Total:", percentual_total)

fig, axs = plt.subplots(1, 3, figsize=(14, 5))

def plot_pizza(ax, data, titulo):
    labels = ['Vitórias', 'Derrotas']
    valores = [data['W'], data['L']]
    cores = ['#4CAF50', '#F44336']  # verde, vermelho
    ax.pie(valores, labels=labels, autopct='%1.1f%%', startangle=90, colors=cores)
    ax.set_title(titulo)

plot_pizza(axs[0], percentual_player, "Player")
plot_pizza(axs[1], percentual_bot, "Bot")
plot_pizza(axs[2], percentual_total, "Total")


plt.suptitle("Percentual de Vitórias e Derrotas")
plt.tight_layout()
qtd_player = len(tabela[tabela['tipo'] == 'Player'])
qtd_bot = len(tabela[tabela['tipo'] == 'Bot'])
qtd_total = len(tabela)

texto_quantidades = (
    f"Player: {qtd_player} partidas\n"
    f"Bot: {qtd_bot} partidas\n"
    f"Total: {qtd_total} partidas"
)
plt.figtext(0.5, -0.05, texto_quantidades, ha='center', fontsize=12)

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

def extract_goals_assists(stat):
    if stat == '0' or stat == 0 or pd.isna(stat):
        return 0, 0

    goals = 0
    assists = 0

    if 'G' in str(stat):
        g_parts = str(stat).split('G')
        goals = int(g_parts[0]) if g_parts[0] else 0

        if len(g_parts) > 1 and 'A' in g_parts[1]:
            a_parts = g_parts[1].split('A')
            assists = int(a_parts[0]) if a_parts[0] else 0

    elif 'A' in str(stat):
        a_parts = str(stat).split('A')
        assists = int(a_parts[0]) if a_parts[0] else 0

    return goals, assists

jogadores = ['Afonso', 'Davi', 'Guilherme']
tipos = ['Bot', 'Player']

gols_por_jogador_tipo = {jogador: {tipo: 0 for tipo in tipos} for jogador in jogadores}
assists_por_jogador_tipo = {jogador: {tipo: 0 for tipo in tipos} for jogador in jogadores}

for idx, row in tabela.iterrows():
    tipo_partida = row['tipo']
    for jogador in jogadores:
        stat = row[jogador]
        gols, assists = extract_goals_assists(stat)
        gols_por_jogador_tipo[jogador][tipo_partida] += gols
        assists_por_jogador_tipo[jogador][tipo_partida] += assists

jogadores_nomes = list(gols_por_jogador_tipo.keys())
gols_bot = [gols_por_jogador_tipo[j]['Bot'] for j in jogadores_nomes]
gols_player = [gols_por_jogador_tipo[j]['Player'] for j in jogadores_nomes]
assists_bot = [assists_por_jogador_tipo[j]['Bot'] for j in jogadores_nomes]
assists_player = [assists_por_jogador_tipo[j]['Player'] for j in jogadores_nomes]

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6))

bar_width = 0.35
x_indexes = np.arange(len(jogadores_nomes))

ax1.bar(x_indexes - bar_width/2, gols_bot, bar_width, label='Contra Bots', color='lightcoral', alpha=0.8)
ax1.bar(x_indexes + bar_width/2, gols_player, bar_width, label='Contra Players', color='coral', alpha=0.8)
ax1.set_title('Gols por Jogador - Partidas vs Bots vs Players', fontsize=14, fontweight='bold')
ax1.set_xlabel('Jogadores', fontweight='bold')
ax1.set_ylabel('Quantidade de Gols', fontweight='bold')
ax1.set_xticks(x_indexes)
ax1.set_xticklabels(jogadores_nomes)
ax1.legend()
ax1.grid(axis='y', alpha=0.3)

for i, v in enumerate(gols_bot):
    ax1.text(i - bar_width/2, v + 0.1, str(v), ha='center', va='bottom', fontweight='bold')
for i, v in enumerate(gols_player):
    ax1.text(i + bar_width/2, v + 0.1, str(v), ha='center', va='bottom', fontweight='bold')

ax2.bar(x_indexes - bar_width/2, assists_bot, bar_width, label='Contra Bots', color='lightblue', alpha=0.8)
ax2.bar(x_indexes + bar_width/2, assists_player, bar_width, label='Contra Players', color='dodgerblue', alpha=0.8)
ax2.set_title('Assistências por Jogador - Partidas vs Bots vs Players', fontsize=14, fontweight='bold')
ax2.set_xlabel('Jogadores', fontweight='bold')
ax2.set_ylabel('Quantidade de Assistências', fontweight='bold')
ax2.set_xticks(x_indexes)
ax2.set_xticklabels(jogadores_nomes)
ax2.legend()
ax2.grid(axis='y', alpha=0.3)

for i, v in enumerate(assists_bot):
    ax2.text(i - bar_width/2, v + 0.1, str(v), ha='center', va='bottom', fontweight='bold')
for i, v in enumerate(assists_player):
    ax2.text(i + bar_width/2, v + 0.1, str(v), ha='center', va='bottom', fontweight='bold')

plt.tight_layout()
plt.show()

print("\n=== RESUMO DE GOLS ===")
for jogador in jogadores_nomes:
    total_gols = gols_por_jogador_tipo[jogador]['Bot'] + gols_por_jogador_tipo[jogador]['Player']
    print(f"{jogador}: {total_gols} gols total")
    print(f"  - Contra Bots: {gols_por_jogador_tipo[jogador]['Bot']} gols")
    print(f"  - Contra Players: {gols_por_jogador_tipo[jogador]['Player']} gols")

print("\n=== RESUMO DE ASSISTÊNCIAS ===")
for jogador in jogadores_nomes:
    total_assists = assists_por_jogador_tipo[jogador]['Bot'] + assists_por_jogador_tipo[jogador]['Player']
    print(f"{jogador}: {total_assists} assistências total")
    print(f"  - Contra Bots: {assists_por_jogador_tipo[jogador]['Bot']} assistências")
    print(f"  - Contra Players: {assists_por_jogador_tipo[jogador]['Player']} assistências")

print("\n=== ESTATÍSTICAS GERAIS ===")
total_partidas_bot = len(tabela[tabela['tipo'] == 'Bot'])
total_partidas_player = len(tabela[tabela['tipo'] == 'Player'])
print(f"Total de partidas contra Bots: {total_partidas_bot}")
print(f"Total de partidas contra Players: {total_partidas_player}")
```

```python
import matplotlib.pyplot as plt

def extract_goals_assists(stat):
    if stat == '0' or stat == 0 or pd.isna(stat):
        return 0, 0

    goals = 0
    assists = 0

    if 'G' in str(stat):
        g_parts = str(stat).split('G')
        goals = int(g_parts[0]) if g_parts[0] else 0

        if len(g_parts) > 1 and 'A' in g_parts[1]:
            a_parts = g_parts[1].split('A')
            assists = int(a_parts[0]) if a_parts[0] else 0

    elif 'A' in str(stat):
        a_parts = str(stat).split('A')
        assists = int(a_parts[0]) if a_parts[0] else 0

    return goals, assists

jogadores = ['Afonso', 'Davi', 'Guilherme']

gols_por_jogador_tipo = {jogador: {'Bot': 0, 'Player': 0} for jogador in jogadores}
assists_por_jogador_tipo = {jogador: {'Bot': 0, 'Player': 0} for jogador in jogadores}

for idx, row in tabela.iterrows():
    tipo_partida = row['tipo']
    for jogador in jogadores:
        stat = row[jogador]
        gols, assists = extract_goals_assists(stat)
        gols_por_jogador_tipo[jogador][tipo_partida] += gols
        assists_por_jogador_tipo[jogador][tipo_partida] += assists

tabela_player = pd.DataFrame({
    'Jogador': jogadores,
    'Gols': [gols_por_jogador_tipo[j]['Player'] for j in jogadores],
    'Assistências': [assists_por_jogador_tipo[j]['Player'] for j in jogadores],
    'Total (G+A)': [gols_por_jogador_tipo[j]['Player'] + assists_por_jogador_tipo[j]['Player'] for j in jogadores]
})

tabela_bot = pd.DataFrame({
    'Jogador': jogadores,
    'Gols': [gols_por_jogador_tipo[j]['Bot'] for j in jogadores],
    'Assistências': [assists_por_jogador_tipo[j]['Bot'] for j in jogadores],
    'Total (G+A)': [gols_por_jogador_tipo[j]['Bot'] + assists_por_jogador_tipo[j]['Bot'] for j in jogadores]
})

total_partidas_player = len(tabela[tabela['tipo'] == 'Player'])
total_partidas_bot = len(tabela[tabela['tipo'] == 'Bot'])

tabela_player['Média G/P'] = tabela_player['Gols'] / total_partidas_player
tabela_player['Média A/P'] = tabela_player['Assistências'] / total_partidas_player
tabela_player['Média T/P'] = tabela_player['Total (G+A)'] / total_partidas_player

tabela_bot['Média G/P'] = tabela_bot['Gols'] / total_partidas_bot
tabela_bot['Média A/P'] = tabela_bot['Assistências'] / total_partidas_bot
tabela_bot['Média T/P'] = tabela_bot['Total (G+A)'] / total_partidas_bot

tabela_player = tabela_player.round(2)
tabela_bot = tabela_bot.round(2)

print("=" * 80)
print("TABELA - PARTIDAS CONTRA PLAYERS")
print("=" * 80)
print(tabela_player.to_string(index=False))
print(f"\nTotal de partidas contra Players: {total_partidas_player}")

print("\n" + "=" * 80)
print("TABELA - PARTIDAS CONTRA BOTS")
print("=" * 80)
print(tabela_bot.to_string(index=False))
print(f"\nTotal de partidas contra Bots: {total_partidas_bot}")

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 6))

ax1.axis('tight')
ax1.axis('off')
tabela_player_display = tabela_player.copy()
tabela_player_display = tabela_player_display.round(2)
table1 = ax1.table(cellText=tabela_player_display.values,
                  colLabels=tabela_player_display.columns,
                  cellLoc='center',
                  loc='center',
                  bbox=[0, 0, 1, 1])
table1.auto_set_font_size(False)
table1.set_fontsize(10)
table1.scale(1.2, 1.5)
ax1.set_title('Estatísticas - Partidas contra Players', fontsize=14, fontweight='bold', pad=20)

ax2.axis('tight')
ax2.axis('off')
tabela_bot_display = tabela_bot.copy()
tabela_bot_display = tabela_bot_display.round(2)
table2 = ax2.table(cellText=tabela_bot_display.values,
                  colLabels=tabela_bot_display.columns,
                  cellLoc='center',
                  loc='center',
                  bbox=[0, 0, 1, 1])
table2.auto_set_font_size(False)
table2.set_fontsize(10)
table2.scale(1.2, 1.5)
ax2.set_title('Estatísticas - Partidas contra Bots', fontsize=14, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()

tabela_player.to_csv('estatisticas_players.csv', index=False)
tabela_bot.to_csv('estatisticas_bots.csv', index=False)
print("\nTabelas salvas como 'estatisticas_players.csv' e 'estatisticas_bots.csv'")
print("\n" + "=" * 80)
print("COMPARAÇÃO ENTRE TIPOS DE PARTIDA")
print("=" * 80)
for jogador in jogadores:
    gols_player = gols_por_jogador_tipo[jogador]['Player']
    gols_bot = gols_por_jogador_tipo[jogador]['Bot']
    assists_player = assists_por_jogador_tipo[jogador]['Player']
    assists_bot = assists_por_jogador_tipo[jogador]['Bot']

    print(f"\n{jogador}:")
    print(f"  Gols: {gols_player} (Players) vs {gols_bot} (Bots)")
    print(f"  Assistências: {assists_player} (Players) vs {assists_bot} (Bots)")
    print(f"  Total: {gols_player + assists_player} (Players) vs {gols_bot + assists_bot} (Bots)")
```
